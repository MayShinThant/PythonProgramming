```python
import pandas as pd
```

```python
df = pd.read_csv("Course Evaluation_Parami.csv")
```

```python
# Checking the null

df.isna().sum()
```

```python
# Check for duplicates
print(df.duplicated().sum())

# Remove duplicates
df_cleaned = df.drop_duplicates()
```

```python
# Checking and listing the column names that are longer than 12 characters

long_columns = [col for col in df.columns if len(col) >= 12]
short_columns = [col for col in df.columns if len(col) < 12]

print(long_columns)

print(len(df.columns))
print(len(long_columns))
print(len(short_columns))
```

```python
# Creating a list with new column names

new_columns = [
    'Attendance_Percentage',
    'Challenge_Expectation',
    'Effort_Course',
    'Preparedness_Class',
    'Performance_Level',
    'Above_Average_Comparison',
    'Teaching_Clarity',
    'Stimulated_Student_Interest',
    'Effective_Time_Use',
    'Instructor_Availability',
    'Instructor_Caring',
    'Increased_Interest',
    'Confidence_Gained',
    'Importance_of_Learning',
    'Course_Challenge',
    'Skill_Development',
    'Study_Hours_Per_Week',
    'Recommendation',
    'Reason'
]

# Renaming the old columns with new columns

column_mapping = dict(zip(long_columns, new_columns))
df.rename(columns=column_mapping, inplace=True)
```

```python
# Removing the outliers and replacing with null value

age_outlier = df['Age'] > 26
study_hours_outliers = df['Study_Hours_Per_Week']>40

df.loc[age_outlier,'Age'] = pd.NA
df.loc[study_hours_outliers,'Study_Hours_Per_Week'] = pd.NA

print(df[['Age','Study_Hours_Per_Week']])
```

```python
# Replacing the null values with mean values

df[['Age','Study_Hours_Per_Week']]=df[['Age','Study_Hours_Per_Week']].fillna(df[['Age','Study_Hours_Per_Week']].mean())
print(df[['Age','Study_Hours_Per_Week']])

df['Age'] = df['Age'].astype(int)
```

```python
# Convert all text in the 'Reason'column to lowercase and remove the full-stop
df['Reason'] = df['Reason'].str.strip()
df['Reason'] = df['Reason'].str.lower()
df['Reason'] = df['Reason'].str.replace('.','', regex = False)
print(df['Reason'])
```

```python
# Describing all the features whether the data type is object

print(df.describe(include='all'))
```

```python
# Standardize categorical values in 'Gender' column
df['Gender'].replace({'Male': 'M', 'Female': 'F'}, inplace=True)
df['Gender']
```

```python
# Mapping categorical to numerical values
mapping = {'strongly agree': 5, 'agree': 4, 'neutral': 3, 'disagree': 2, 'strongly disagree': 1}

# Apply mapping to relevant columns
columns_to_map = [
    'Challenge_Expectation',
    'Effort_Course',
    'Preparedness_Class',
    'Performance_Level',
    'Above_Average_Comparison',
    'Teaching_Clarity',
    'Stimulated_Student_Interest',
    'Effective_Time_Use',
    'Instructor_Availability',
    'Instructor_Caring',
    'Increased_Interest',
    'Confidence_Gained',
    'Importance_of_Learning',
    'Course_Challenge',
    'Skill_Development'
]

df[columns_to_map] = df[columns_to_map].astype(str).apply(lambda x: x.str.lower())

df[columns_to_map] = df[columns_to_map].applymap(lambda x: mapping.get(x, x))
print(df[columns_to_map])
```

```python
# Looking at the study hours based on the reasons

opportunity_students = df[df['Reason'] == 'to take opportunity']
study_hours_opportunity = opportunity_students['Study_Hours_Per_Week']
print("Study Hours for Students Choosing 'to take opportunity' as the Reason:")
print(study_hours_opportunity)

interest_students = df[df['Reason'] == 'interest']
study_hours_interest = interest_students['Study_Hours_Per_Week']
print("Study Hours for Students Choosing 'interest' as the Reason:")
print(study_hours_interest)

degree_students = df[df['Reason'] == 'degree requirement']
study_hours_degree = degree_students['Study_Hours_Per_Week']
print("Study Hours for Students Choosing 'degree requirement' as the Reason:")
print(study_hours_degree)
```

```python
# Comparing the mean of the students who have more than 7 study hours (median study hours), based on their reasons

high_study_hours_students = df[df['Study_Hours_Per_Week'] > 7]
grouped_means = high_study_hours_students.groupby('Reason').mean()

columns_of_interest = ['Recommendation','Study_Hours_Per_Week','Challenge_Expectation','Effort_Course', 'Preparedness_Class','Performance_Level', 'Above_Average_Comparison']

print("Mean Values for Effort, Preparedness, and Performance for Students with Study Hours > 8:")
print(grouped_means[columns_of_interest])
```
